```python
def get_df(x,y):
    start = list(range(0,min(len(x),len(y)),3))
    stop = list(range(3,min(len(x),len(y)),3))

    df = pd.DataFrame({"start":start[:-1],"stop":stop})
    
    df['codon_wt'] = df.apply(lambda row: x[row['start']:row['stop']], axis=1)
    df['codon_mutant'] = df.apply(lambda row: y[row['start']:row['stop']], axis=1)

    empty_index_wt = df.index[df['codon_wt'].str.contains("X")]
    empty_index_mutant = df.index[df['codon_mutant'].str.contains("X")]

    if len(empty_index_wt) == 0:
        df['aa_wt'] = df.apply(lambda row: dict[row['codon_wt']],axis = 1)

    if len(empty_index_mutant) == 0:
        df['aa_mutant'] = df.apply(lambda row: dict[row['codon_mutant']],axis = 1)
        
    if len(empty_index_mutant) != 0:
        if len(empty_index_mutant) == 1:
            df.loc[empty_index_mutant,'codon_mutant'] = "XXX"
        else:
            empty_df = df.loc[empty_index_mutant]
            scavenged_codon = ''.join(df.loc[empty_index_mutant, 'codon_mutant']).replace("X","")

            df.loc[empty_index_mutant,'codon_mutant'] = "XXX"

            if len(empty_df.loc[df['aa_wt']==dict[scavenged_codon]])==0:
                df.loc[min(empty_index_mutant),'codon_mutant'] = scavenged_codon
            else:
                ind = empty_df.index[empty_df['aa_wt']==dict[scavenged_codon]]
                df.loc[empty_index_wt, 'codon_mutant'] = "XXX"
                df.loc[ind, 'codon_mutant'] = scavenged_codon
        df['aa_mutant'] = df.apply(lambda row: dict[row['codon_mutant']],axis = 1)

    if len(empty_index_wt) != 0:
        if len(empty_index_wt) == 1:
            df.loc[empty_index_wt,'codon_wt'] = "XXX"
        else:
            empty_df = df.loc[empty_index_wt]
            scavenged_codon = ''.join(df.loc[empty_index_wt, 'codon_wt']).replace("X","")
            empty_df['replace'] = empty_df['codon_wt'].str.replace("X","")
            # contains_bps_index = empty_df.index[empty_df['replace'].str.len()>0]

            df.loc[empty_index_mutant,'codon_wt'] = "XXX"

            if len(empty_df.loc[df['aa_mutant']==dict[scavenged_codon]])==0:
                df.loc[min(empty_index_mutant),'aa_mutant'] = scavenged_codon
            else:
                codon_ind = empty_df.index[empty_df['aa_mutant']==dict[scavenged_codon]]
                df.loc[empty_index_wt, 'codon_wt'] = "XXX"
                df.loc[codon_ind, 'codon_wt'] = scavenged_codon
        # print(df.loc[empty_index_wt])
        df['aa_wt'] = df.apply(lambda row: dict[row['codon_wt']],axis = 1)


    wt_length = min(df.index[df['aa_wt'] == '*'])+1
    mutant_length = min(df.index[df['aa_mutant'] == '*'])+1

    df = df[:max(wt_length,mutant_length)]

    return(df)

```

```python
def mutate(original, new, start, stop = 0):
    
    coding_wt = nuc[len(not_translated):]

    if stop == 0:
        stop = start

    coding_mutant = ''.join([coding_wt[:start-1],new,coding_wt[stop:]])
    full_mutant = ''.join([not_translated,coding_mutant])
    print(full_mutant)
    
    if abs(len(coding_wt)-len(coding_mutant)) == 0 : # Missense - if substituting something that is the same length 
        if coding_wt[start-1:stop] != original:
            print("Error in entry")
            print("Expected bps = ", coding_wt[start-1:stop])
        else:
            df = get_df(coding_wt, coding_mutant)
            index = df.index[df['aa_mutant']!=df['aa_wt']]

            if len(index) > 1:
                if len(''.join(df.loc[index,'aa_mutant'])) > 0:
                    name = [''.join(df.loc[index,'aa_wt']),str(min(index)+1),"_",str(max(index)+1),"del_ins_", ''.join(df.loc[index,'aa_mutant'])]
                    print("Deletion/Insertion/Missense mutation: ", ''.join(name))
            else:
                name = [df.loc[min(index),'aa_wt'],str(min(index)+1),df.loc[min(index),'aa_mutant']]
                if df.loc[min(index),'aa_mutant'] == "*":
                    print("Nonsense mutation: ", ''.join(name))
                else:
                    print("Missense mutation: ", ''.join(name))
                
            print(''.join(df['aa_mutant']))
    elif abs(len(coding_mutant)-len(coding_wt)) % 3 == 0: # no frameshift
        if coding_wt[start-1:stop] != original:
            print("Error in entry")
            print("Expected bps = ", coding_wt[start-1:stop])
        else:    
            if len(original) > len(new): # deletion
                deletion = len(original)*"X"
                print(deletion)
                coding_mutant_same_length = ''.join([coding_wt[:start-1],deletion,coding_wt[stop:]]) # will be the same as the original dataframe but some will be missing (b/t "start" and "stop")
                
                # print(coding_mutant_same_length[start-5:start+5])
                df = get_df(coding_wt, coding_mutant_same_length)
                index = df.index[df['aa_mutant']!=df['aa_wt']]

                if len(index) > 1:
                    if len(''.join(df.loc[index,'aa_mutant'])) > 0:
                        name = [''.join(df.loc[index,'aa_wt']),str(min(index)+1),"_",str(max(index)+1),"del_ins_", ''.join(df.loc[index,'aa_mutant'])]
                    else:
                        name = [''.join(df.loc[index,'aa_wt']),str(min(index)+1),"_",str(max(index)+1),"del"]
                else:
                    name = [''.join(df.loc[index,'aa_wt']),str(min(index)+1),"del"]
                print(''.join(df['aa_mutant']))
                print("Deletion mutation: ", ''.join(name))
            else: # insertion/duplication 
                insertion = (len(new)-1)*"X"
                coding_wt_same_length = ''.join([coding_wt[:start-1],original, insertion,coding_wt[stop:]])
                df = get_df(coding_wt_same_length, coding_mutant)

                index = df.index[df['aa_mutant']!=df['aa_wt']]
                ins = ''.join(df.loc[index,'aa_mutant'])

                before_index = index - len(index)
                before = ''.join(df.loc[before_index,'aa_wt'])
                after_index  = index + len(index)
                after = ''.join(df.loc[after_index,'aa_wt'])

                if ins == before or ins == after:
                    print(pro.index(ins))
                    if len(ins) > 1: 
                        name = [''.join(df.loc[index,'aa_wt']),str(min(index)+1),"_",str(max(index)+1),"dup"]
                    else:
                        name = [''.join(df.loc[index,'aa_wt']),str(min(index)+1),"dup"]
                else:
                    if len(ins) > 1: 
                        name = [''.join(df.loc[index,'aa_wt']),str(min(index)+1),"_",str(max(index)+1),"ins"]
                    else:
                        name = [''.join(df.loc[index,'aa_wt']),str(min(index)+1),"ins"]
                # print(''.join(df['aa_mutant']))

                # print(pro.index(ins))
                
                # name = [''.join(df.loc[index,'aa_mutant']),str(min(index)+1),"ins"]
                name = ''.join(name)
                
                print("Insertion mutation:", name)
                
    else: # frameshift
        if coding_wt[start-1:stop] != original:
            print("Error in entry")
            print("Expected bps = ", coding_wt[start-1:stop])
        else:    
            df = get_df(coding_wt, coding_mutant)
            start_index = min(df.index[df['aa_mutant']!=df['aa_wt']])
            ter_index = min(df.index[df['aa_mutant'] == '*'])
            
            frameshift = ter_index - start_index 

            name = [df.loc[start_index,'aa_wt'],str(start_index+1),df.loc[start_index,'aa_mutant'],"fs*",str(frameshift+1)]
            print(''.join(df['aa_mutant']))
            print("Frameshift mutation: ", ''.join(name))

    
    image_index = df.index[df['aa_mutant']!=df['aa_wt']]
    # print(image_index)
    # print(df)
    df['star'] = " "
    df.loc[image_index, 'star'] = "*"
    df.loc[df['aa_wt']=="",'aa_wt'] = ' '
    df.loc[df['aa_mutant']=="",'aa_mutant'] = ' '
    min_image_index = max(min(image_index)-5,min(df.index))
    max_image_index = min(max(image_index)+6,max(df.index))
    image_df = df.iloc[min_image_index:max_image_index]

    print("")
    if min_image_index < 9:
        print(' ' + '   '.join(list(map(str, range(min_image_index, min(max_image_index,9),1)))),
            ' ' + '  '.join(list(map(str, range(max(min_image_index,10), min(max_image_index,99),1)))),
            ' '.join(list(map(str, range(max(min_image_index,100), max_image_index,1)))))
    elif min_image_index < 99:
        print(' ' + '  '.join(list(map(str, range(max(min_image_index,10), min(max_image_index,99),1)))),
            ' '.join(list(map(str, range(max(min_image_index,100), max_image_index,1)))))
    else:
         print(' '.join(list(map(str, range(max(min_image_index,100), max_image_index,1)))))
    print("wt")
    print(' '.join(image_df['codon_wt']))
    print(' ' + '   '.join(image_df['aa_wt']))
    print(' ' + '   '.join(image_df['star']))
    print(' ' + '   '.join(image_df['aa_mutant']))
    print(' '.join(image_df['codon_mutant']))
    print("mutant")

    # row = int(start/3)
    # ind = df.iloc[row-5:row+5]
    # print(ind)
        
```
